```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/dataset/감성대화말뭉치(최종데이터)_Training.xlsx')
```

```python
# '사람문장1' 컬럼을 'text'로 이름 변경
df.rename(columns={'사람문장1': 'text'}, inplace=True)

# '감정_대분류' 컬럼을 'emotion'으로 이름 변경
df.rename(columns={'감정_대분류': 'emotion'}, inplace=True)

df = df[['text', 'emotion']]

print(df)
```

```python
# 각 감정별로 카운트
emotion_counts = df['emotion'].value_counts()

print(emotion_counts)
```

```python
# '상처' 감정 제거
df = df[df['emotion'] != '상처']
```

```python
year4 = pd.read_csv('/content/drive/MyDrive/dataset/4차년도.csv', encoding='cp949')
year5 = pd.read_csv('/content/drive/MyDrive/dataset/5차년도.csv', encoding='cp949')
year5_2 = pd.read_csv('/content/drive/MyDrive/dataset/5차년도_2차.csv', encoding='cp949')

# 열 이름 변경
year4 = year4.rename(columns={"4번감정세기": "4번 감정세기"})
# 열 이름 변경
year5 = year5.rename(columns={"4번감정세기": "4번 감정세기"})
# 열 이름 변경
year5_2 = year5_2.rename(columns={"4번감정세기": "4번 감정세기"})
```

```python
year4 = year4[year4['상황'] == 'neutral']
year5 = year5[year5['상황'] == 'neutral']
year5_2 = year5_2[year5_2['상황'] == 'neutral']

netural_df = pd.concat([year4, year5, year5_2])
df = df[['text', 'emotion']]
netural_df = netural_df[['발화문', '상황']]

netural_df.rename(columns={'발화문': 'text'}, inplace=True)
netural_df.rename(columns={'상황': 'emotion'}, inplace=True)

netural_df = netural_df[netural_df['text'].str.len() >= 5]
```

```python
netural_df
```

```python
df = pd.concat([df, netural_df])

df
```

```python
# 각 감정별로 카운트
df['emotion'] = df['emotion'].replace('neutral', '중립')
emotion_counts = df['emotion'].value_counts()

print(emotion_counts)
```

```python
df.to_csv('/content/drive/MyDrive/dataset/감성대화_대화음성.csv')
```
