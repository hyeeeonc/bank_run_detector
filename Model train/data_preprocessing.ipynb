```python
import pandas as pd

year4 = pd.read_csv('/content/drive/MyDrive/dataset/4차년도.csv', encoding='cp949')
year5 = pd.read_csv('/content/drive/MyDrive/dataset/5차년도.csv', encoding='cp949')
year5_2 = pd.read_csv('/content/drive/MyDrive/dataset/5차년도_2차.csv', encoding='cp949')

# 열 이름 변경
year4 = year4.rename(columns={"4번감정세기": "4번 감정세기"})
# 열 이름 변경
year5 = year5.rename(columns={"4번감정세기": "4번 감정세기"})
# 열 이름 변경
year5_2 = year5_2.rename(columns={"4번감정세기": "4번 감정세기"})
```

```python
# 중복 제거 전 개수 출력
print("중복 제거 전")
print("year4:", year4.shape[0])
print("year5:", year5.shape[0])
print("year5_2:", year5_2.shape[0])

year4 = year4.drop_duplicates(subset=["발화문"])
year5 = year5.drop_duplicates(subset=["발화문"])
year5_2 = year5_2.drop_duplicates(subset=["발화문"])

# 중복 제거 후 개수 출력
print("중복 제거 후")
print("year4:", year4.shape[0])
print("year5:", year5.shape[0])
print("year5_2:", year5_2.shape[0])
```

```python
# 2. 각자 감정 카운팅해서 사용하기

df = year5_2

# 감정 세기를 계산하여 새로운 열에 저장합니다.
emotions = ["angry", "happiness", "disgust", "fear", "neutral", "sadness", "surprise"]

for emotion in emotions:
    df[emotion] = 0

for index, row in df.iterrows():
    for i in range(1, 6):
        emotion_col = f"{i}번 감정"
        intensity_col = f"{i}번 감정세기"
        # print(k, emotion_col, intensity_col)
        emotion = row[emotion_col]
        intensity = row[intensity_col]
        # print(k, emotion, intensity)
        if emotion in emotions:
            df.at[index, emotion] += intensity
            # print("?????", emotion, intensity, df.at[index, emotion])
        else:
            print(emotion, "탈락")
        # 각 행의 상태를 출력합니다.
    # print(f"Index: {index}, Row: {row}")

# 가장 큰 감정 세기를 가진 감정을 찾아 해당 열에 저장합니다.
df["emotion"] = df[emotions].idxmax(axis=1)

# 만약 최대 감정 세기를 가진 감정이 두 개 이상이면 상황을 고려하여 다시 결정합니다.
for index, row in df.iterrows():
    max_emotion = row["emotion"]
    max_intensity = row[max_emotion]
    max_intensity_count = sum(row[emotions] == max_intensity)
    if max_intensity_count > 1:
        situation_emotion = row["상황"]
        if situation_emotion in emotions:
            df.at[index, "emotion"] = situation_emotion
        else:
            print("상황 틀림", situation_emotion)

print(df.head())
```
